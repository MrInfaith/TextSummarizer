preprocess the data

#### remove punctuations

```python
import string 
punc=string.punctuation
```

```python
def remove_punc(text):
    return text.translate(str.maketrans('','',punc))
```

```python
text=list(map(lambda s:remove_punc(s),text))
```

```python
text
```

##### lower

```python
text=list(map(lambda s:s.lower(),text))
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stopwords=stopwords.words('english')
```

```python
text = [word for word in text if word not in stopwords]
```

```python
len(text)
```

```python
word_frequencies={}
for word in text:
    if word not in word_frequencies.keys():
        word_frequencies[word]=1
    else:
        word_frequencies[word]+=1
```

```python
word_frequencies
```

```python
max_frequency=max(word_frequencies.values())
```

```python
max_frequency
```

```python
#normalized
for word in word_frequencies.keys():
    word_frequencies[word]=word_frequencies[word]/max_frequency
```

```python
#sentences formations 
sentences = ""
for page_num in range(len(a.pages)):
    page = a.pages[page_num]
    sentences+= page.extract_text()
```

```python
sentences
```

```python
sentences=sentences.replace('\n','8055')
```

```python
sentences=sentences.replace(' ','')
```

```python
sentences=sentences.replace('8055',' ')
```

```python
sentences=sentences.split(".")
```

```python
sentences
```

```python
sentences_scores={}
for sent in sentences:
    for word in sent:
        if word.lower() in word_frequencies.keys():
            if sent not in sentences_scores.keys():
                 sentences_scores[sent]=word_frequencies[word.lower()]
            else:
                 sentences_scores[sent]+=word_frequencies[word.lower()]
```

```python
sentences_scores
```

```python
from heapq import nlargest
```

```python
select_length=int(len(sentences)*0.35)
```

```python
summary=nlargest(select_length,sentences_scores,key=sentences_scores.get)
```

```python
summary
```

```python
final_summary=[word for word in summary]
```

```python
final_summary='.'.join(final_summary)
```
